Semi-Isomorphism Test

```python
import networkx as nx

def is_semi_isomorphism(A1, A2):
    
    """"
    Tests if two DFAs A1 = (Q1, Σ1, τ1, (qs)1, F1) and A2 = (Q2, Σ2, τ2, (qs)2, F2) 
    are semi-isomorphic, for this to occur if there exists two bijections:
    1) f: Q1 → Q2 and 2) g: Σ1 → Σ2

    where 
    τ2(f(q), g(a)) = f(τ1(q,a)) for all q ∈ Q1 and all a ∈ Σ1
    """

    """
    Input for the function is two DFAs (A1, A2) and function will return a boolean output
    (True or False) depending on whether the two DFAs ar semi-isomorphic.

    Q is some set of finite states (python type - set)
    Σ is some finite alphabet (python type - set)
    τ is a transition function τ: Q x Σ → Q (python type - dict)
    qs is a start state (python type - int)
    F ⊆ Q is a set of final states (python type - set)
    """
    # Obtaining values from input

    Q1, sigma1, tau1, qs1, F1 = A1
    Q2, sigma2, tau2, qs2, F2 = A2

    # Checking if |Q1| = |Q2| and |Σ1| = |Σ2|
    # If the state sets and alphabets have different sizes no bijection can exist
    if len(Q1) != len(Q2) or len(sigma1) != len(sigma2):
        return False, None, None

    # Convert AFDs to a directed graph on NetworkX for easier use
    G1 = nx.MultiDiGraph()
    for q1 in Q1:
        G1.add_node(q1)
        for a1 in sigma1:
            next_state = tau1[(q1, a1)]
            G1.add_node(next_state)
            G1.add_edge(q1, next_state, label = 1)


    G2 = nx.MultiDiGraph()
    for q2 in Q2:
        G2.add_node(q2)
        for a2 in sigma2:
            next_state = tau2[(q2, a2)]
            G2.add_node(next_state)
            G2.add_edge(q2, next_state, label = 2)

    # We assume that in each graph there is one state/node from which all other nodes are reachable.
    # From the SCCC functionality in NetworkX, in one single SCC, properties are inherited since they are all connected,
    # therefore, if there is one SCC which can reach every other SCC our node is in this SCC
    # We can then draw an semi-isomorphisms between these SCCs

    sccs1 = list(nx.strongly_connected_components(G1))
    sccs2 = list(nx.strongly_connected_components(G2))

    # Obtain a condensation of G1 which is the graph of the SCCs

    C1 = nx.condensation(G1, scc = sccs1)
    C2 = nx.condensation(G2, scc = sccs2)
    reaching_states1 = []
    reaching_states2 = []

    for scc in C1:
        if C1.in_degree(scc) == 0:
            reaching_states1.extend(sccs1[scc])

    for scc in C2:
        if C2.in_degree(scc) == 0:
            reaching_states2.extend(sccs2[scc])

    # We set up an isomorphism between the sets of states which reach every other state
    # for both A1 and A2, and that is the root of the semi-isomorphism.
    # From this we take pairs of (root1, root2) where root1 is a state
    # which can reach every state of A1 and similarly for root2.
    # We try to establish f(root1) = root2 as an initial mapping,
    # we then try to establish the rest of f and g by continuing the previous transitions.
    # If no combination is possible then no semi-isomorphism exists.
    # Since it is a semi-isomorphism we don't care about starting states, hence
    # the initial mapping of f isn't determined by qs — we try all initial root pairings.
    
    for root1 in reaching_states1:
        for root2 in reaching_states2:

            # Initial mapping, we know one mapping from the SCC
            # g starts empty as we don't know anything about the alphabet mapping yet
            f = {root1: root2}
            g = {}

            queue = [root1]
            visited = {root1}
            check = True

            for q in queue:
                fq = f[q]
                for a in sigma1:
                    q_real = tau1[(q, a)] # where A1 goes from q on symbol a

                    if a in g and q_real in f: # Case 1: we already know both g(a) and f(q_real)
                        if tau2[(fq, g[a])] != f[q_real]:
                            check = False
                            break

                    elif a in g and q_real not in f: # Case 2: we know g(a) but not f(q_real)
                        target = tau2[(fq, g[a])]
                        if target in f.values():
                            check = False
                            break
                        f[q_real] = target
                        if q_real not in visited:
                            queue.append(q_real)
                            visited.add(q_real)
                                
                    elif a not in g and q_real in f: #  Case we know f(q_real) but not g(a)
                        found = False
                        for b in sigma2:
                            if b not in g.values() and tau2[(fq, b)] == f[q_real]:
                                g[a] = b
                                found = True
                                break
                        if not found:
                            check = False
                            break
                            
                    else: # Case 4: we know neither g(a) nor f(q_real)
                        found = False
                        for b in sigma2:
                            if b not in g.values() and tau2[(fq, b)] not in f.values():
                                f[q_real] = tau2[(fq, b)]
                                g[a]       = b
                                if q_real not in visited:
                                    queue.append(q_real)
                                    visited.add(q_real)
                                found = True
                                break
                        if not found:
                            check = False
                            break

                if not check:
                    break

            if check and len(f) == len(Q1) and len(g) == len(sigma1):
                return True, f, g

    return False, None, None
```

```python
import matplotlib.pyplot as plt
def visualise(A1, A2, results):
    Q1, sigma1, tau1, qs1, F1 = A1
    Q2, sigma2, tau2, qs2, F2 = A2

    # Convert AFDs to a directed graph on NetworkX for easier use
    G1 = nx.MultiDiGraph()
    for q1 in Q1:
        G1.add_node(q1)
        for a1 in sigma1:
            next_state = tau1[(q1, a1)]
            G1.add_node(next_state)
            G1.add_edge(q1, next_state, label = a1)

    G2 = nx.MultiDiGraph()
    for q2 in Q2:
        G2.add_node(q2)
        for a2 in sigma2:
            next_state = tau2[(q2, a2)]
            G2.add_node(next_state)
            G2.add_edge(q2, next_state, label = a2)

    # Set up the two side by side plots
    fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 5))

    # Calculate positions for the nodes
    pos1 = nx.circular_layout(G1)             # ← you were missing these two lines
    pos2 = nx.circular_layout(G2)

    # Draw both graphs
    nx.draw(G1, pos1, ax=ax1, with_labels=True, node_color='lightblue',
            node_size=1500, arrows=True)
    nx.draw(G2, pos2, ax=ax2, with_labels=True, node_color='lightblue',
            node_size=1500, arrows=True)

    ax1.set_title("A1")
    ax2.set_title("A2")

    # Show result in the main title
    found, f, g = result
    if found:
        fig.suptitle(f"SEMI-ISOMORPHIC\nf={f}\ng={g}", color='green')
    else:
        fig.suptitle("NOT SEMI-ISOMORPHIC", color='red')

    plt.show()
```

In part (c) you should assume that in each input automata there is at least one state from which all other states are reachable. You may find the functionality in NetworkX for Strongly Connected Components and Condensations useful for identifying such states.

For isomorphism, you can generate isomorphic DFAs by picking one DFA and applying a random permutation to the states; if you then change a value in the transition function you will likely get non-isomorphic DFAs.
